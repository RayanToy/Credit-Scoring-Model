## Замечания по интерпретации WOE Encoding
- В отличие от one-hot encoding, где каждая категория становится отдельной переменной, WOE Encoding преобразует каждую категориальную переменную в одну числовую переменную, значение которой отражает логарифм отношения шансов "хорошего" исхода (не дефолт) к "плохому" исходу (дефолт) для каждой категории. Положительные значения WOE указывают на категории с более высоким риском дефолта, а отрицательные — с более низким.
- Для точной интерпретации влияния каждой категории (например, "мужчины" vs "женщины" для переменной "ПОЛ") нужно посмотреть на значения WOE, которые были рассчитаны для каждой категории. Эти значения можно получить из объекта `woe_encoder`, если вы хотите детализировать анализ.

Для числовых признаков применяется RobustScaler - метод нормализации, устойчивый к выбросам, который использует медиану и межквартильный размах вместо среднего значения и стандартного отклонения. Категориальные переменные обрабатываются с помощью Weight of Evidence (WOE) кодирования - это стандартный подход в кредитном скоринге, который преобразует категории в числовые значения, пропорциональные их связи с целевой переменной, сохраняя при этом монотонную зависимость.

Оптимизация гиперпараметров моделей выполняется с использованием библиотеки Optuna, которая оптимизирует модель в пространстве гиперпараметров, максимизируя метрику AUC-ROC. Эта метрика выбрана как основная, поскольку она не зависит от порога классификации и хорошо отражает способность модели к ранжированию клиентов. Для каждой обученной модели мы дополнительно определяем оптимальный порог классификации, перебирая значения от 0.1 до 0.9 и выбирая то, которое дает максимальное значение Gini-коэффициента (производной от AUC-ROC).

Особое внимание уделено обработке категориальных признаков: для большинства моделей используется WOE-кодирование, в то время как для CatBoost сохраняются исходные категории, поскольку этот алгоритм имеет встроенные эффективные механизмы работы с ними. Весь процесс завершается комплексной оценкой моделей, сравнением их метрик (AUC-ROC, Gini, F1) и анализом матриц ошибок, что позволяет выбрать наилучшую модель как с точки зрения точности, так и интерпретируемости.

```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import (roc_auc_score, classification_report,
                           f1_score, roc_curve, auc, confusion_matrix)
from category_encoders import WOEEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import optuna
from optuna.samplers import TPESampler
from functools import partial

# Функция для расчета Gini коэффициента
def gini_score(y_true, y_pred_proba):
    return 2 * roc_auc_score(y_true, y_pred_proba) - 1

# Функция для поиска оптимального порога классификации
def find_optimal_threshold(y_true, y_pred_proba):
    thresholds = np.linspace(0.1, 0.9, 50)  # Тестируем 50 порогов от 0.1 до 0.9
    best_threshold = 0.5  # Начальное значение порога
    best_gini = -1  # Начальное значение Gini

    # Перебираем пороги и находим тот, который дает максимальный Gini
    for threshold in thresholds:
        y_pred = (y_pred_proba >= threshold).astype(int)
        current_gini = gini_score(y_true, y_pred)
        if current_gini > best_gini:
            best_gini = current_gini
            best_threshold = threshold

    return best_threshold

# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y)  # 30% тестовых данных с сохранением распределения y

# Определение категориальных и числовых признаков
categorical_columns = [
    'ПОЛ', 'СМЕНА_МЖ', 'ЗАГРАН', 'БРАК', 'ДЕТИ', 'ИЖДИВЕНЦЫ', 'ОБРАЗОВАНИЕ',
    'ИНОСТР.ЯЗЫК', 'АВТО', 'ОТРАСЛЬ', 'РАБОТА_ПО_НАПР', 'РАБОТА_В_ОРГ',
    'ДОЛЖНОСТЬ', 'ОТДЕЛЕНИЕ', 'ГОРОД', 'СОБСТВЕННИК_ФАКТ.', 'КРЕДИТЫ_ПОГАШЕНЫ',
    'ПЕРВЫЙ_КРЕДИТ'
]

numeric_columns = [
    'ВОЗРАСТ', 'ДОХОД_СЕМЬИ_', 'ДОХОД_НА_ОДНОГО_ЧЛЕНА_СЕМЬИ',
    'ПЕРСОНАЛЬНЫЙ_ДОХОД', 'СУММА_ВЫДАННОГО_КРЕДИТА',
    'ДОХОД_СУПРУГИ(А)'
]

# Кодирование категориальных переменных с помощью WOE
woe_encoder = WOEEncoder(cols=categorical_columns)
woe_encoder.fit(X_train, y_train)

X_train_woe = woe_encoder.transform(X_train)
X_test_woe = woe_encoder.transform(X_test)

# Масштабирование числовых признаков
scaler = RobustScaler()  # Устойчивый к выбросам scaler
X_train_woe[numeric_columns] = scaler.fit_transform(X_train_woe[numeric_columns])
X_test_woe[numeric_columns] = scaler.transform(X_test_woe[numeric_columns])

# Подготовка данных для CatBoost (без WOE-кодирования)
X_train_cat = X_train.copy()
X_test_cat = X_test.copy()
X_train_cat[numeric_columns] = scaler.fit_transform(X_train_cat[numeric_columns])
X_test_cat[numeric_columns] = scaler.transform(X_test_cat[numeric_columns])

# Функция для оптимизации гиперпараметров моделей с помощью Optuna
def optimize_model(trial, model_name, X, y, cat_features=None):
    params = {}

    # Оптимизация параметров для логистической регрессии
    if model_name == 'Logistic Regression':
        penalty_solver_combinations = [
            'l1:liblinear',
            'l1:saga',
            'l2:newton-cg',
            'l2:lbfgs',
            'l2:liblinear',
            'l2:sag',
            'l2:saga',
            'elasticnet:saga'
        ]

        penalty_solver = trial.suggest_categorical('penalty_solver_str', penalty_solver_combinations)
        penalty, solver = penalty_solver.split(':')

        params = {
            'C': trial.suggest_float('C', 1e-3, 10, log=True),
            'penalty': penalty,
            'solver': solver,
            'max_iter': trial.suggest_int('max_iter', 100, 1000)
        }

        if penalty == 'elasticnet':
            params['l1_ratio'] = trial.suggest_float('l1_ratio', 0, 1)

        model = LogisticRegression(class_weight='balanced', **params)

    # Оптимизация параметров для случайного леса
    elif model_name == 'Random Forest':
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 50, 500),
            'max_depth': trial.suggest_int('max_depth', 3, 20),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 20),
            'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
            'bootstrap': trial.suggest_categorical('bootstrap', [True, False])
        }
        model = RandomForestClassifier(class_weight='balanced', random_state=42, **params)

    # Оптимизация параметров для XGBoost
    elif model_name == 'XGBoost':
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 50, 500),
            'max_depth': trial.suggest_int('max_depth', 3, 10),
            'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.3, log=True),
            'subsample': trial.suggest_float('subsample', 0.5, 1),
            'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
            'gamma': trial.suggest_float('gamma', 0, 5),
            'reg_alpha': trial.suggest_float('reg_alpha', 0, 10),
            'reg_lambda': trial.suggest_float('reg_lambda', 0, 10),
            'tree_method': trial.suggest_categorical('tree_method', ['exact', 'approx', 'hist'])
        }
        model = XGBClassifier(
            scale_pos_weight=len(y[y==0])/len(y[y==1]),  # Балансировка классов
            random_state=42,
            **params
        )

    # Оптимизация параметров для CatBoost
    elif model_name == 'CatBoost':
        params = {
            'iterations': trial.suggest_int('iterations', 50, 500),
            'depth': trial.suggest_int('depth', 3, 10),
            'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.3, log=True),
            'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
            'border_count': trial.suggest_int('border_count', 32, 255),
            'random_strength': trial.suggest_float('random_strength', 1e-3, 10, log=True),
            'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise', 'Lossguide']),
            'bootstrap_type': trial.suggest_categorical('bootstrap_type', ['Bayesian', 'Bernoulli', 'MVS']),
            'leaf_estimation_method': trial.suggest_categorical('leaf_estimation_method', ['Newton', 'Gradient']),
            'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations', 1, 10)
        }
        if params['bootstrap_type'] == 'Bayesian':
            params['bagging_temperature'] = trial.suggest_float('bagging_temperature', 0, 10)
        elif params['bootstrap_type'] == 'Bernoulli':
            params['subsample'] = trial.suggest_float('subsample', 0.5, 1)

        model = CatBoostClassifier(
            cat_features=cat_features,  # Указание категориальных признаков
            auto_class_weights='Balanced',  # Автоматическая балансировка классов
            random_state=42,
            verbose=0,  # Отключение вывода обучения
            **params
        )

    # Оптимизация параметров для линейного дискриминантного анализа
    elif model_name == 'Linear Discriminant':
        solver = trial.suggest_categorical('solver', ['svd', 'lsqr', 'eigen'])
        params = {
            'solver': solver,
            'tol': trial.suggest_float('tol', 1e-5, 1e-2, log=True)
        }
        if solver in ['lsqr', 'eigen']:
            params['shrinkage'] = trial.suggest_float('shrinkage', 0, 1)

        model = LinearDiscriminantAnalysis(**params)

    # Оценка модели с кросс-валидацией
    try:
        score = np.mean(cross_val_score(model, X, y, cv=5, scoring='roc_auc', error_score='raise'))
        return score
    except Exception as e:
        print(f"Error with params {params}: {str(e)}")
        raise optuna.TrialPruned()  # Пропуск неудачных комбинаций параметров

# Список моделей для оптимизации
models_to_optimize = [
    'Logistic Regression',
    'Random Forest',
    'XGBoost',
    'CatBoost',
    'Linear Discriminant'
]

# Функция для обучения финальной модели с лучшими параметрами
def train_final_model(model_name, params, X_train, y_train):
    if model_name == 'Logistic Regression':
        penalty, solver = params['penalty_solver_str'].split(':')
        model_params = {
            'C': params['C'],
            'penalty': penalty,
            'solver': solver,
            'max_iter': params['max_iter'],
            'class_weight': 'balanced'
        }
        if penalty == 'elasticnet' and 'l1_ratio' in params:
            model_params['l1_ratio'] = params['l1_ratio']
        model = LogisticRegression(**model_params)

    elif model_name == 'Random Forest':
        model = RandomForestClassifier(
            class_weight='balanced',
            random_state=42,
            **params
        )

    elif model_name == 'XGBoost':
        model = XGBClassifier(
            scale_pos_weight=len(y_train[y_train==0])/len(y_train[y_train==1]),
            random_state=42,
            **params
        )

    elif model_name == 'CatBoost':
        model = CatBoostClassifier(
            cat_features=categorical_columns,
            auto_class_weights='Balanced',
            random_state=42,
            verbose=0,
            **params
        )

    elif model_name == 'Linear Discriminant':
        model = LinearDiscriminantAnalysis(**params)

    model.fit(X_train, y_train)
    return model

# Оптимизация и обучение моделей
best_params = {}  # Словарь для хранения лучших параметров
final_models = {}  # Словарь для хранения обученных моделей
thresholds = {}    # Словарь для хранения оптимальных порогов
n_trials = 100     # Количество испытаний для оптимизации каждой модели

for model_name in models_to_optimize:
    print(f"\nOptimizing {model_name}...")

    # Создание исследования Optuna
    study = optuna.create_study(direction='maximize', sampler=TPESampler(seed=42))

    # Выбор данных для обучения (CatBoost использует свои данные)
    if model_name == 'CatBoost':
        opt_func = partial(optimize_model, model_name=model_name, X=X_train_cat, y=y_train, cat_features=categorical_columns)
    else:
        opt_func = partial(optimize_model, model_name=model_name, X=X_train_woe, y=y_train)

    # Запуск оптимизации
    study.optimize(opt_func, n_trials=n_trials)

    # Сохранение лучших параметров
    best_params[model_name] = study.best_params
    print(f"Best params for {model_name}: {study.best_params}")
    print(f"Best AUC: {study.best_value:.4f}")

    # Обучение финальной модели
    print(f"\nTraining final {model_name}...")
    if model_name == 'CatBoost':
        final_model = train_final_model(model_name, study.best_params, X_train_cat, y_train)
    else:
        final_model = train_final_model(model_name, study.best_params, X_train_woe, y_train)

    final_models[model_name] = final_model

    # Определение оптимального порога классификации
    X_test_data = X_test_cat if model_name == 'CatBoost' else X_test_woe
    y_pred_proba = final_model.predict_proba(X_test_data)[:, 1]
    optimal_threshold = find_optimal_threshold(y_test, y_pred_proba)
    thresholds[model_name] = optimal_threshold
    print(f"Optimal threshold for {model_name}: {optimal_threshold:.4f}")

# Функция для оценки модели с кросс-валидацией
def evaluate_model_with_cv(model, X, y, model_name, categorical_columns=None, n_folds=5):
    skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)
    auc_scores, gini_scores, f1_scores, thresholds_cv = [], [], [], []
    all_conf_matrices = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
        X_train_fold, X_val_fold = X.iloc[train_idx], X.iloc[val_idx]
        y_train_fold, y_val_fold = y.iloc[train_idx], y.iloc[val_idx]

        # Обработка данных для CatBoost
        if model_name == 'CatBoost':
            X_train_fold_cat = X_train_fold.copy()
            X_val_fold_cat = X_val_fold.copy()
            scaler = RobustScaler()
            X_train_fold_cat[numeric_columns] = scaler.fit_transform(X_train_fold_cat[numeric_columns])
            X_val_fold_cat[numeric_columns] = scaler.transform(X_val_fold_cat[numeric_columns])
            model.fit(X_train_fold_cat, y_train_fold, cat_features=categorical_columns)
            y_pred_proba = model.predict_proba(X_val_fold_cat)[:, 1]
        else:
            # WOE-кодирование и масштабирование для других моделей
            woe_encoder = WOEEncoder(cols=categorical_columns)
            X_train_fold_woe = woe_encoder.fit_transform(X_train_fold, y_train_fold)
            X_val_fold_woe = woe_encoder.transform(X_val_fold)

            scaler = RobustScaler()
            X_train_fold_woe[numeric_columns] = scaler.fit_transform(X_train_fold_woe[numeric_columns])
            X_val_fold_woe[numeric_columns] = scaler.transform(X_val_fold_woe[numeric_columns])

            model.fit(X_train_fold_woe, y_train_fold)
            y_pred_proba = model.predict_proba(X_val_fold_woe)[:, 1]

        # Поиск оптимального порога для текущего фолда
        optimal_threshold = find_optimal_threshold(y_val_fold, y_pred_proba)
        thresholds_cv.append(optimal_threshold)

        # Прогнозирование и расчет метрик
        y_pred = (y_pred_proba >= optimal_threshold).astype(int)

        auc_roc = roc_auc_score(y_val_fold, y_pred_proba)
        gini = gini_score(y_val_fold, y_pred_proba)
        f1 = f1_score(y_val_fold, y_pred)
        conf_matrix = confusion_matrix(y_val_fold, y_pred)

        # Сохранение метрик
        auc_scores.append(auc_roc)
        gini_scores.append(gini)
        f1_scores.append(f1)
        all_conf_matrices.append(conf_matrix)

    # Расчет средних значений метрик
    mean_auc = np.mean(auc_scores)
    mean_gini = np.mean(gini_scores)
    mean_f1 = np.mean(f1_scores)
    mean_threshold = np.mean(thresholds_cv)
    mean_conf_matrix = np.mean(all_conf_matrices, axis=0).astype(int)

    return {
        'Mean AUC-ROC': mean_auc,
        'Mean Gini': mean_gini,
        'Mean F1': mean_f1,
        'Mean Threshold': mean_threshold,
        'Mean Confusion Matrix': mean_conf_matrix,
        'AUC-ROC per fold': auc_scores,
        'Gini per fold': gini_scores,
        'F1 per fold': f1_scores
    }

# Оценка моделей с кросс-валидацией
cv_results = {}
for name, model in final_models.items():
    print(f"\nEvaluating {name} with cross-validation...")
    if name == 'CatBoost':
        cv_results[name] = evaluate_model_with_cv(model, X, y, name, categorical_columns=categorical_columns)
    else:
        cv_results[name] = evaluate_model_with_cv(model, X, y, name, categorical_columns=categorical_columns)

    # Вывод результатов кросс-валидации
    print(f"\nResults for {name} (Cross-Validation):")
    print(f"Mean AUC-ROC: {cv_results[name]['Mean AUC-ROC']:.4f}")
    print(f"Mean Gini: {cv_results[name]['Mean Gini']:.4f}")
    print(f"Mean F1: {cv_results[name]['Mean F1']:.4f}")
    print(f"Mean Threshold: {cv_results[name]['Mean Threshold']:.4f}")
    print("Mean Confusion Matrix:")
    print(cv_results[name]['Mean Confusion Matrix'])

# Выбор лучшей модели по AUC-ROC
best_model_name = max(cv_results, key=lambda x: cv_results[x]['Mean AUC-ROC'])
best_model = final_models[best_model_name]
best_threshold = cv_results[best_model_name]['Mean Threshold']

print(f"\nBest model: {best_model_name}")

# Оценка лучшей модели на тестовой выборке
X_test_data = X_test_cat if best_model_name == 'CatBoost' else X_test_woe
y_pred_proba_test = best_model.predict_proba(X_test_data)[:, 1]
y_pred_test = (y_pred_proba_test >= best_threshold).astype(int)

# Расчет метрик на тестовой выборке
auc_roc_test = roc_auc_score(y_test, y_pred_proba_test)
gini_test = gini_score(y_test, y_pred_proba_test)
f1_test = f1_score(y_test, y_pred_test)
conf_matrix_test = confusion_matrix(y_test, y_pred_test)

# Вывод результатов для тестовой выборки
print(f"\nFinal evaluation on test set for {best_model_name}:")
print(f"AUC-ROC: {auc_roc_test:.4f}")
print(f"Gini: {gini_test:.4f}")
print(f"F1-score: {f1_test:.4f}")
print(f"Threshold: {best_threshold:.4f}")
print("Classification Report:")
print(classification_report(y_test, y_pred_test))
print("Confusion Matrix:")
print(conf_matrix_test)

# Сравнение всех моделей на тестовой выборке
results = {}
plt.figure(figsize=(12, 10))

for name, model in final_models.items():
    X_test_data = X_test_cat if name == 'CatBoost' else X_test_woe

    # Прогнозирование вероятностей и классов
    y_pred_proba = model.predict_proba(X_test_data)[:, 1]
    y_pred = (y_pred_proba >= thresholds[name]).astype(int)

    # Расчет метрик
    auc_roc = roc_auc_score(y_test, y_pred_proba)
    gini = gini_score(y_test, y_pred_proba)
    f1 = f1_score(y_test, y_pred)

    # Сохранение результатов
    results[name] = {
        'AUC-ROC': auc_roc,
        'Gini': gini,
        'F1': f1,
        'Threshold': thresholds[name],
        'Confusion Matrix': confusion_matrix(y_test, y_pred)
    }

    # Построение ROC-кривой
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f}, Threshold = {thresholds[name]:.2f})')

    # Вывод метрик для текущей модели
    print(f"\n{name}:")
    print(f"AUC-ROC: {auc_roc:.4f}")
    print(f"Gini: {gini:.4f}")
    print(f"F1-score: {f1:.4f}")
    print(f"Optimal threshold: {thresholds[name]:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

# Визуализация ROC-кривых
plt.plot([0, 1], [0, 1], 'k--')  # Линия случайного классификатора
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves Comparison with Optimal Thresholds')
plt.legend(loc="lower right", bbox_to_anchor=(1.4, 0))
plt.show()

# Сводная таблица метрик всех моделей
results_df = pd.DataFrame(results).T
print("\nСводная таблица метрик всех моделей:")
print(results_df.sort_values(by='AUC-ROC', ascending=False))
```
